# 读取“星表”

```python
def read_minorplanet(flie_name):
    with open(flie_name,"r") as f :
        linenumber = 0
        star_dde = []
        the_dde = dict()
        for line in f.readlines():
            linenumber += 1
            if linenumber % 2 == 0:
                the_dde.update({'dde': line})
                star_dde.append(the_dde.copy())
                the_dde.clear()
    return star_dde
```

```python
import ephem
import math
start_time = '2022/7/20 13:07:48'
end_time = '2022/7/20 22:39:25'
ob = ephem.Observer()
ob.lon, ob.lat = '93.20', '38.45'
ob.date = start_time
aa = ephem.date(end_time)
obsever_start =ob.next_setting(ephem.Sun()) # 太阳落下的时间
observer_end =ob.next_rising(ephem.Sun())#太阳升起得时间
step_time = ephem.second*4
```

# 得到所有的恒星在可观测夜晚的alt

```python
star_list1 = [] #用来收集所有的恒星在整个夜晚的alt

# bb = ephem.date(start_time)
for i in range(50):
    observer_start = ephem.date(start_time)
    obeject_information = [] #用来收集每颗恒星在整个夜晚的alt
    while observer_start<observer_end:
        ob.date = observer_start
        star = ephem.readdb(c[i]["dde"])
        star.compute(ob)
        alt = star.alt
        
        obeject_information.append(math.degrees(alt))
        observer_start+=step_time
    star_list1.append(obeject_information)
```
